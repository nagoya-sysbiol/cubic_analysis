# Figure 7e and Supplementary Figure 10a

##Plot minimum distance of between each points

In this tutrial, we provide two figure (Fig7e and SupFig10a).

Firstly, we load required packages and data.

```
library(readr)
library(ggplot2)
system("wget https://www.dropbox.com/s/3rokwbw0o7iggtr/fig7e.zip")
system("unzip /content/fig7e.zip")
distdf = read_csv("/content/Fig7e.csv")
```

```
head(distdf)
```

Fig7e is violin plots. We use this plot to see overall trends.

```
ggplot(distdf, aes(x=factor(day), y=log10(1+dist), group=day))+
  geom_violin()+
  geom_boxplot(width=0.3)+
  theme_bw(14)+
  xlab("day")
```

SupFig10a is a combined of the histograms. We use this plot to show more detail of the data.

```
ggplot(distdf,aes(x=log10(1+dist)))+
  geom_histogram()+
  facet_wrap(~day+id,scales = "free_y",
             labeller = label_context)+
  theme_bw(14)
```
